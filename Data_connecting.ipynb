```python
import gc
import pandas as pd
```

```python
data_path = "./data/"
```

# Data reading

```python
AdsInfo = pd.read_csv(data_path + "AdsInfo.tsv", sep = '\t')
```

```python
AdsInfo.head()
```

```python
AdsInfo.info()
```

```python
Category = pd.read_csv(data_path + "Category.tsv", sep = '\t')
```

```python
Category.head()
```

```python
Category.info()
```

```python
Location = pd.read_csv(data_path + "Location.tsv", sep = '\t')
```

```python
Location.head()
```

```python
Location.info()
```

```python
trainSearchStream = pd.read_csv(data_path + "trainSearchStream.tsv", sep = '\t').sample(10000000)
```

```python
trainSearchStream = trainSearchStream[~trainSearchStream['IsClick'].isnull()]
```

```python
trainSearchStream.head()
```

```python
trainSearchStream.info()
```

# Connecting all together

```python
%%time
data = pd.merge(trainSearchStream, AdsInfo, how = 'left', on = "AdID")
data.head()
```

```python
%%time
data = pd.merge(data, Category, how = 'left', on = "CategoryID")
data.head()
```

```python
%%time
data = pd.merge(data, Location, how = 'left', on = "LocationID")
data.head()
```

```python
data = data.rename(columns = {'Level_x' : 'Level'})
```

```python
data.to_parquet(data_path + "merged_data" + ".parquet")
```
